1. $f(x, y) = (x, 0)$

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de transformacin
M = np.array([[1, 0], [0, 0]])

# Generar vectores aleatorios de norma 1 y 3
V = np.random.rand(2, 10) - 0.5
V_uni = V / np.linalg.norm(V, axis=0)
V_3 = 3 * V / np.linalg.norm(V, axis=0)

# Aplicar transformacin
MV_uni = M @ V_uni
MV_3 = M @ V_3

# Crear figura ms compacta
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4))
fig.suptitle('Transformacin $f(x, y) = (x, 0)$', fontsize=12)

# Funcin auxiliar para graficar flechitas
def draw_vectors(ax, original, transformed, color1='blue', color2='red'):
    for i in range(original.shape[1]):
        ax.quiver(0, 0, original[0, i], original[1, i], angles='xy', scale_units='xy', scale=1, color=color1, alpha=0.7)
        ax.quiver(0, 0, transformed[0, i], transformed[1, i], angles='xy', scale_units='xy', scale=1, color=color2, alpha=0.7)

# Grfico para vectores de norma 1
draw_vectors(ax1, V_uni, MV_uni)
ax1.set_title('Radio 1')
ax1.set_aspect('equal')
ax1.set_xlim(-1.2, 1.2)
ax1.set_ylim(-1.2, 1.2)
ax1.grid(True, linestyle='--', alpha=0.4)

# Grfico para vectores de norma 3
draw_vectors(ax2, V_3, MV_3)
ax2.set_title('Radio 3')
ax2.set_aspect('equal')
ax2.set_xlim(-3.2, 3.2)
ax2.set_ylim(-3.2, 3.2)
ax2.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```

2. $f(x, y) = (x,y)$

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de reflexin respecto al eje x
M = np.array([[1, 0], [0, -1]])

# Puntos aleatorios normalizados
V = np.random.rand(2, 2) - 0.5
V_uni = V / np.linalg.norm(V, axis=0)
V_3 = 3 * V / np.linalg.norm(V, axis=0)

# Transformacin
MV_uni = M @ V_uni
MV_3 = M @ V_3

# Grfico con referencias
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

for i in range(V_uni.shape[1]):
    ax1.quiver(0, 0, V_uni[0, i], V_uni[1, i], angles='xy', scale_units='xy', scale=1, color='blue')
    ax1.quiver(0, 0, MV_uni[0, i], MV_uni[1, i], angles='xy', scale_units='xy', scale=1, color='red')

for i in range(V_3.shape[1]):
    ax2.quiver(0, 0, V_3[0, i], V_3[1, i], angles='xy', scale_units='xy', scale=1, color='blue')
    ax2.quiver(0, 0, MV_3[0, i], MV_3[1, i], angles='xy', scale_units='xy', scale=1, color='red')

# Detalles y ejes
for ax in (ax1, ax2):
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_aspect('equal')
    ax.grid(True)

ax1.set_title('Radio 1')
ax2.set_title('Radio 3')
fig.legend(['Original', 'Transformado'], loc='lower center', ncol=2)
fig.suptitle('Reflexin con f(x, y) = (x, -y)', fontsize=14)
plt.tight_layout()
plt.show()
```

3. $f(x, y) = ( \frac{1}{2} (x + y), \frac{1}{2} (x + y))$

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de proyeccin sobre la recta y = x
M = np.array([[0.5, 0.5], [0.5, 0.5]])

# Puntos aleatorios normalizados
V = np.random.rand(2, 2) - 0.5
V_uni = V / np.linalg.norm(V, axis=0)
V_3 = 3 * V / np.linalg.norm(V, axis=0)

# Transformacin
MV_uni = M @ V_uni
MV_3 = M @ V_3

# Grfico con referencias
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

for i in range(V_uni.shape[1]):
    ax1.quiver(0, 0, V_uni[0, i], V_uni[1, i], angles='xy', scale_units='xy', scale=1, color='blue')
    ax1.quiver(0, 0, MV_uni[0, i], MV_uni[1, i], angles='xy', scale_units='xy', scale=1, color='red')

for i in range(V_3.shape[1]):
    ax2.quiver(0, 0, V_3[0, i], V_3[1, i], angles='xy', scale_units='xy', scale=1, color='blue')
    ax2.quiver(0, 0, MV_3[0, i], MV_3[1, i], angles='xy', scale_units='xy', scale=1, color='red')

# Detalles y ejes
for ax in (ax1, ax2):
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_aspect('equal')
    ax.grid(True, linestyle='--', alpha=0.5)

    # Lneas punteadas de referencia
    x_vals = np.linspace(-4, 4, 100)
    ax.plot(x_vals, x_vals, '--', color='gray', linewidth=1, label='y = x')

ax1.set_title('Radio 1')
ax2.set_title('Radio 3')
fig.legend(['Original', 'Transformado'], loc='lower center', ncol=4)
fig.suptitle('Proyeccin sobre y = x: f(x, y) = ((x+y), (x+y))', fontsize=14)
plt.tight_layout()
plt.show()
```

4. $f(x, y) = (x\ast cos(t)  y\ast sen(t) , x\ast sen (t) + y\ast cos (t))$

```python
import numpy as np
import matplotlib.pyplot as plt

# ngulo de rotacin (45 grados en radianes)
t = np.pi / 4

# Matriz de rotacin
M = np.array([
    [np.cos(t), -np.sin(t)],
    [np.sin(t),  np.cos(t)]
])

# Puntos aleatorios normalizados
V = np.random.rand(2, 1) - 0.5
V_uni = V / np.linalg.norm(V, axis=0)
V_3 = 3 * V / np.linalg.norm(V, axis=0)

# Transformacin
MV_uni = M @ V_uni
MV_3 = M @ V_3

# Grfico con referencias
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

for i in range(V_uni.shape[1]):
    ax1.quiver(0, 0, V_uni[0, i], V_uni[1, i], angles='xy', scale_units='xy', scale=1, color='blue')
    ax1.quiver(0, 0, MV_uni[0, i], MV_uni[1, i], angles='xy', scale_units='xy', scale=1, color='red')

for i in range(V_3.shape[1]):
    ax2.quiver(0, 0, V_3[0, i], V_3[1, i], angles='xy', scale_units='xy', scale=1, color='blue')
    ax2.quiver(0, 0, MV_3[0, i], MV_3[1, i], angles='xy', scale_units='xy', scale=1, color='red')

# Detalles y ejes
for ax in (ax1, ax2):
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_aspect('equal')
    ax.grid(True, linestyle='--', alpha=0.5)

ax1.set_title('Radio 1')
ax2.set_title('Radio 3')
fig.legend(['Original', 'Transformado'], loc='lower center', ncol=2)
fig.suptitle(r'Rotacin antihoraria: $f(x, y) = (x\cos t - y\sin t,\ x\sin t + y\cos t)$', fontsize=14)
plt.tight_layout()
plt.show()
```
