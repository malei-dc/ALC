1. $f(x, y) = (x, 0)$

````python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de transformación
M = np.array([[1, 0], [0, 0]])

# Generar vectores aleatorios de norma 1 y 3
V = np.random.rand(2, 10) - 0.5
V_uni = V / np.linalg.norm(V, axis=0)
V_3 = 3 * V / np.linalg.norm(V, axis=0)

# Aplicar transformación
MV_uni = M @ V_uni
MV_3 = M @ V_3

# Crear figura más compacta
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4))
fig.suptitle('Transformación $f(x, y) = (x, 0)$', fontsize=12)

# Función auxiliar para graficar flechitas
def draw_vectors(ax, original, transformed, color1='blue', color2='red'):
    for i in range(original.shape[1]):
        ax.quiver(0, 0, original[0, i], original[1, i], angles='xy', scale_units='xy', scale=1, color=color1, alpha=0.7)
        ax.quiver(0, 0, transformed[0, i], transformed[1, i], angles='xy', scale_units='xy', scale=1, color=color2, alpha=0.7)

# Gráfico para vectores de norma 1
draw_vectors(ax1, V_uni, MV_uni)
ax1.set_title('Radio 1')
ax1.set_aspect('equal')
ax1.set_xlim(-1.2, 1.2)
ax1.set_ylim(-1.2, 1.2)
ax1.grid(True, linestyle='--', alpha=0.4)

# Gráfico para vectores de norma 3
draw_vectors(ax2, V_3, MV_3)
ax2.set_title('Radio 3')
ax2.set_aspect('equal')
ax2.set_xlim(-3.2, 3.2)
ax2.set_ylim(-3.2, 3.2)
ax2.grid(True, linestyle='--', alpha=0.4)

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
````

2. $f(x, y) = (x,−y)$

````python
import numpy as np
import matplotlib.pyplot as plt

# Matriz de reflexión respecto al eje x
M = np.array([[1, 0], [0, -1]])

# Puntos aleatorios normalizados
V = np.random.rand(2, 2) - 0.5
V_uni = V / np.linalg.norm(V, axis=0)
V_3 = 3 * V / np.linalg.norm(V, axis=0)

# Transformación
MV_uni = M @ V_uni
MV_3 = M @ V_3

# Gráfico con referencias
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

for i in range(V_uni.shape[1]):
    ax1.quiver(0, 0, V_uni[0, i], V_uni[1, i], angles='xy', scale_units='xy', scale=1, color='blue')
    ax1.quiver(0, 0, MV_uni[0, i], MV_uni[1, i], angles='xy', scale_units='xy', scale=1, color='red')

for i in range(V_3.shape[1]):
    ax2.quiver(0, 0, V_3[0, i], V_3[1, i], angles='xy', scale_units='xy', scale=1, color='blue')
    ax2.quiver(0, 0, MV_3[0, i], MV_3[1, i], angles='xy', scale_units='xy', scale=1, color='red')

# Detalles y ejes
for ax in (ax1, ax2):
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_aspect('equal')
    ax.grid(True)

ax1.set_title('Radio 1')
ax2.set_title('Radio 3')
fig.legend(['Original', 'Transformado'], loc='lower center', ncol=2)
fig.suptitle('Reflexión con f(x, y) = (x, -y)', fontsize=14)
plt.tight_layout()
plt.show()
````

4. $f(x, y) = (x\ast cos(t) − y\ast sen(t) , x\ast sen (t) + y\ast cos (t))$

````python
import numpy as np
import matplotlib.pyplot as plt

# Ángulo de rotación (45 grados en radianes)
t = np.pi / 4

# Matriz de rotación
M = np.array([
    [np.cos(t), -np.sin(t)],
    [np.sin(t),  np.cos(t)]
])

# Puntos aleatorios normalizados
V = np.random.rand(2, 1) - 0.5
V_uni = V / np.linalg.norm(V, axis=0)
V_3 = 3 * V / np.linalg.norm(V, axis=0)

# Transformación
MV_uni = M @ V_uni
MV_3 = M @ V_3

# Gráfico con referencias
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

for i in range(V_uni.shape[1]):
    ax1.quiver(0, 0, V_uni[0, i], V_uni[1, i], angles='xy', scale_units='xy', scale=1, color='blue')
    ax1.quiver(0, 0, MV_uni[0, i], MV_uni[1, i], angles='xy', scale_units='xy', scale=1, color='red')

for i in range(V_3.shape[1]):
    ax2.quiver(0, 0, V_3[0, i], V_3[1, i], angles='xy', scale_units='xy', scale=1, color='blue')
    ax2.quiver(0, 0, MV_3[0, i], MV_3[1, i], angles='xy', scale_units='xy', scale=1, color='red')

# Detalles y ejes
for ax in (ax1, ax2):
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_aspect('equal')
    ax.grid(True, linestyle='--', alpha=0.5)

ax1.set_title('Radio 1')
ax2.set_title('Radio 3')
fig.legend(['Original', 'Transformado'], loc='lower center', ncol=2)
fig.suptitle(r'Rotación antihoraria: $f(x, y) = (x\cos t - y\sin t,\ x\sin t + y\cos t)$', fontsize=14)
plt.tight_layout()
plt.show()
````

12. $sen(10^j \pi)$ para $1\leq j \leq 25$

````python

for j in range(1, 26):
    x = 10**j * np.pi
    s = np.sin(x)
    print(f"j = {j:2d}, sin(10^{j} * π) = {s:.20f}")
````

>  matemáticamente debería ser $0$ para todos los valores de $j$, ya que $sin(𝑛\pi)=0$ para cualquier entero $n$.
>
> Pero, ¿qué está pasando?
>
> - Para valores pequeños de $j$, el resultado va a estar muy cerca de cero.
> - Pero a medida que 10j crece, va perdiendo precisión en la representación del número, porque no puede representar exactamente números tan grandes multiplicados por $\pi$.
> - Así, el argumento del $sin$ ya no es exactamente múltiplo de $\pi$, y $sin(...)$ puede empezar a dar valores no tan cercanos a $0$.

## Ejercicio 18:

Se quiere estimar la norma-2 de una matriz $A \in \mathbb{R}^{n \times n}$ como el máximo del valor $∥Ax∥{\scriptsize 2}/∥x∥{\scriptsize 2}$ entre varios vectores $x \in \mathbb{R}^3$ no nulos generados al azar. Hacer un programa que
reciba una matriz $A$ y luego

- genere los primeros 100 términos de la siguiente sucesión:

    $$s{\scriptsize 1}=0, s{\scriptsize k+1} = max[s_k, \frac{\|Ax{\scriptsize k}\|{\scriptsize 2}}{\|x{\scriptsize k}\|{\scriptsize 2}}] $$

    donde los $x_k \in \mathbb{R}^3$ son vectores no nulos generados al azar en la bola unitaria: $B = [x :∥x∥_2 ≤ 1]$.

- grafique la sucesión calculada, junto con el valor exacto de la norma de la matriz.

Recordar que tanto la norma-2 puede calcularse con el comando ```np.linalg.norm```. Tener en cuenta que los vectores generados al azar (comando ```np.random.random```) tienen coordenadas en el intervalo [0, 1].

````python
import numpy as np
import matplotlib.pyplot as plt

def estimar_norma2(A, pasos=100):
    n = A.shape[0]
    s = [0]  # sucesión s1 = 0

    for _ in range(pasos):
        # Generar un vector aleatorio en la bola unitaria de R^n
        x = np.random.randn(n)  # muestra de una normal estándar
        x /= np.linalg.norm(x)  # normaliza para que quede en la esfera
        r = np.random.rand() ** (1/n)  # factor radial para estar en la bola unitaria
        x *= r  # escala dentro de la bola unitaria

        # Calcular el cociente de normas
        cociente = np.linalg.norm(A @ x, 2) / np.linalg.norm(x, 2)

        # Actualizar la sucesión
        s.append(max(s[-1], cociente))

    return s

# creo matriz aleatoria
np.random.seed(42)  # para reproducibilidad
A = np.random.rand(3, 3)  # matriz aleatoria de 3x3

# Estimar la norma
s = estimar_norma2(A)
norma_exacta = np.linalg.norm(A, 2)

# Graficar la sucesión y la norma exacta
plt.figure(figsize=(8, 5))
plt.plot(s, label='Estimación de la norma-2')
plt.axhline(y=norma_exacta, color='r', linestyle='--', label='Norma-2 exacta')
plt.xlabel('Iteración')
plt.ylabel('Valor estimado')
plt.title('Estimación de la norma-2 de una matriz')
plt.legend()
plt.grid(True)
plt.show()

print("A = ", A)
print("iteracion 1 de la estimacion:", s[0])
print("iteracion 2 de la estimacion:", s[1])
print("iteracion 3 de la estimacion:", s[2])
print("iteracion 4 de la estimacion:", s[3])
print("...")
print("iteracion 99 de la estimacion:", s[98])
print("norma exacta:", norma_exacta)
````

# Ejercicio 20

Realizar experimentos numéricos para verificar las estimaciones del ítem anterior. Considerar $b = (3, 2, 2)^t$, que se corresponde con la solución exacta $x = (1, 1, 1)^t$. Generar vectores de error aleatorios, normalizarlos para que su norma sea tan chica como la estimada en el item anterior y perturbar $b$ obteniendo $\overset{\sim}{b}$. Finalmente, resolver $A\overset{\sim}{x} = \overset{\sim}{b}$ y verificar que $\|\overset{\sim}{x} − x\| < 10^{−4}$.

````python
import numpy as np

# 1. Definir matriz A y b
A = np.array([
    [3, 0, 0],
    [0, 5/4, 3/4],
    [0, 3/4, 5/4]
])

b = np.array([3, 2, 2])
x_exact = np.array([1, 1, 1])

# 2. Número de condición teórico
cond_infty = 6
max_error_x = 1e-4
max_error_b = max_error_x / cond_infty  # ≈ 1.6667e-5

# 3. Generar vector de error aleatorio con norma infinita 1
np.random.seed(0)  # para reproducibilidad
e = np.random.randn(3)
e = e / np.linalg.norm(e, ord=np.inf)  # normalizar en norma ∞

# 4. Escalar el error para que tenga la norma deseada
perturbation = max_error_b * np.linalg.norm(b, ord=np.inf) * e

# 5. Construir b perturbado
b_perturbed = b + perturbation

# 6. Resolver el sistema con el b perturbado
x_perturbed = np.linalg.solve(A, b_perturbed)

# 7. Calcular error relativo en x
rel_error_x = np.linalg.norm(x_perturbed - x_exact, ord=np.inf) / np.linalg.norm(x_exact, ord=np.inf)

# 8. Mostrar resultados
print(f"Perturbación máxima permitida en b (relativa): {max_error_b:.2e}")
print(f"Perturbación aplicada a b (absoluta): {perturbation}")
print(f"x perturbado: {x_perturbed}")
print(f"Error relativo en x (norma ∞): {rel_error_x:.2e}")
print("¿Se cumple que el error en x < 1e-4?", rel_error_x < 1e-4)
````
